**statsmodel qqplot():**

Can take arguments specifying the parameters for dist or fit them automatically.

Q-Q plot of the quantiles of x versus the quantiles of a distribution.

```python
import numpy as np
import statsmodels.api as sm
import pylab

test = np.random.normal(20,5,1000)  # Draw random samples from a normal (Gaussian) distribution.

sm.qqplot(test, loc=20, scale = 5, line ='45')
pylab.show()
```

If fit is false, loc, scale, and distargs are passed to the distribution. 

If fit is True then the parameters for dist are fit automatically using dist.fit. The quantiles are formed from the standardized data, after subtracting the fitted loc and dividing by the fitted scale.

```python
sm.qqplot(test, loc=20,fit=True, scale = 5, line ='45')
pylab.show()
```

# Explain in your own words what is standard error of the mean?



The standard deviation (SD) measures the amount of variability, or dispersion, for a subject set of data from the mean, 

while the standard error of the mean (SEM) measures how far the sample mean of the data is likely to be from the true population mean. 

The SEM is always smaller than the SD.


The SD describes a dispersion measure of the SAMPLE. You can find roughly 66% of the sample data between ±1 SD for normal distributed variable.

The mean estimated by our sample is usually an estimator of the population mean. But when you draw more than one sample from the same population, the mean values will slightly vary. Therefore, the SE(M) gives us an estimator for the variation. 

SE is calculated : 

$$ SE(M)= \frac{\sigma}{\sqrt{N}}$$

Here

$\sigma$ = Standard deviation

while N is number of samples


With growing N, SEM becomes smaller. That's why the power of statistical tests increases with increasing N, e.g. the difference between two mean values is divided by the standard error of the mean.
